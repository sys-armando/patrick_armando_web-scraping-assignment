```python

import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://maluku.bps.go.id/publication.html"
res = requests.get(url)
```

```python
res.content
```

```python
soup = BeautifulSoup(res.content, 'html.parser')
soup
```

```python
products = soup.find_all("div", class_="pub even col-md-12 col-xs-12 col-sm-12")
products
```

```python
products[0]
```

```python
list_products = []
for product in products[:]:
  item = dict()
  item['Title'] = product.find("div", class_="thumbnail-judul-publikasi").text
  item['Catalog Number'] = product.find("span", class_="thumbnail-nomor-katalog").text
  item['Release Date'] = product.find("span", class_="tanggal-rilis").text
  item['Abstract'] = product.find("span", class_="thumbnail-abstrak-publikasi").text

  list_products.append(item)

df = pd.DataFrame(list_products)
```

```python
df.to_csv('publikasibps.csv', index=False)
```
